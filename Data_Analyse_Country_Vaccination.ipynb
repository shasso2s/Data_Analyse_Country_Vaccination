```python
#the top countries with hightest vaccination count
cols=["country","total_vaccinations"]
df10=data[cols].groupby('country').sum().sort_values("total_vaccinations",ascending=False)
df10=df10.head(10)
df10=df10.sort_values(by="total_vaccinations",ascending=True)
df10=df10.reset_index()
df10
```

```python
plt.plot(df10.total_vaccinations,df10.country,"*--g")
plt.ylabel("country",fontsize="15")
plt.xlabel("total_vaccinations",fontsize="15")
plt.title("countries with highest vaccination count", fontsize="20")
plt.show()
```

```python
plt.barh(df10.country,df10.total_vaccinations)
plt.ylabel("country",fontsize="15")
plt.xlabel("total_vaccinations",fontsize="15")
plt.title("countries with highest vaccination count", fontsize="20")
plt.show()
```

```python
cols=["country","total_vaccinations_per_hundred"]
df11=data[cols].groupby('country').sum().sort_values('total_vaccinations_per_hundred',ascending=False)
df11=df11.head(10)
df11=df11.sort_values(by="total_vaccinations_per_hundred",ascending=True)
df11=df11.reset_index()
df11

```

```python
plt.plot(df11.total_vaccinations_per_hundred,df11.country,"+--g")
plt.xlabel("country")
plt.ylabel("total_vaccinations_per_hundred")
plt.title("countries with hightest vaccination count",fontsize="20")
plt.show()
```

```python
plt.barh(df11.country,df11.total_vaccinations_per_hundred)
plt.xlabel("country")
plt.ylabel("total_vaccinations_per_hundred")
plt.title("countries with hightest vaccination count",fontsize="20")
plt.show()
```
